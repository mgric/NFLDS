# Preliminary Analysis

```python
%matplotlib inline
%pylab inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# import the cleaned dataset into a pandas data frame
df = pd.read_csv('datasets/saintsstatsclean.csv' ,encoding='latin1')
```

I am going to preform some preliminary analysis just to see how our data looks. Now that the data is nice and cleaned up this should be easy. First lets do a real simple sanity check. Lets see how many wins the Saints have had where they score more points than the other team.

```python
wins = df['Result'] == "W"
MP = df['H Score'] > df['O Score']
LP = df['H Score'] < df['O Score']

winNo = df[wins]['Result'].count()
winsMP=df[wins & MP]['Result'].count()
winsLP=df[wins & LP]['Result'].count()

fig, axes = plt.subplots(1, 2, figsize=(16, 4));


labels = ['More Points', 'Fewer Points'];
sizes = [winsMP, winsLP];
colors = [ 'gold', 'lightskyblue'];
axes[0].pie(sizes, labels=labels,colors=colors,autopct='%1.1f%%', shadow=True, startangle=20)
axes[0].axis('equal');
axes[0].set_title('% of Games Won');

labels = ['More Points', 'Fewer Points','Wins']
sizes = [winsMP, winsLP,winNo]
colors = ['yellowgreen', 'gold', 'lightskyblue']
axes[1].bar((0,1,2),sizes,1,color=colors)
axes[1].set_xticks([0.5, 1.5, 2.5])
axes[1].set_xticklabels( labels);
axes[1].set_ylabel('Number of Games')
axes[1].set_title('Number of Games Won');
```

So it looks pretty good. All of the Saints wins have come from games where they scored more points than the other team. While a nice sanity check, they do not really reveal much information. Lets see if the saints win more when they have the home field advantage.

```python
Away = df['Playing'] == "Away"
Home = df['Playing'] == "Home"

winsA=df[wins & Away]['Result'].count()
winsH=df[wins & Home]['Result'].count()

fig, axes = plt.subplots(figsize=(8,4), dpi=100);


labels = ['Away', 'Home'];
sizes = [winsA, winsH];
colors = [ 'gold', 'lightskyblue'];
axes.pie(sizes, labels=labels,colors=colors,autopct='%1.1f%%', shadow=True, startangle=90)
axes.axis('equal');
axes.set_title('% of Games Won');

```

Doesn't look too conclusive. The percentages are too close to say the Saints play better in Louisana then anywhere else. Lets see if attendance plays any part. For this one we will look at losses as well. However, our dataset does not see attendance as numbers, but rather as objects. First we have to change the type

```python

df['Attend']=df['Attend'].str.replace(',','').astype(float)
df[:5]
```

```python
#number of bins
N=4
lossesAT=[]
winsAT=[]
labels=[]

minAt=min(df['Attend'])
maxAt=max(df['Attend'])

binsize1=(maxAt-minAt)/N

wins = df['Result'] == "W"
losses = df['Result'] == "L"

for i in range(0 , N):
    bins=(df['Attend'] >= minAt+i*binsize1) & (df['Attend'] < minAt+(i+1)*binsize1)
    lossesAT.append(df[losses & bins]['Result'].count())
    winsAT.append(df[wins & bins]['Result'].count())
    label = str(minAt+i*binsize1) + " - " + str(minAt+(i+1)*binsize1)
    labels.append(label)
    

ind = np.arange(N)  
width = 0.45

fig, axes = plt.subplots(figsize=(14,4), dpi=100);

lossesAttend = axes.bar(ind, lossesAT, width, color='black')
winsAttend = axes.bar(ind+width, winsAT, width, color='gold')
axes.set_xticks(ind+width); 
axes.set_xticklabels( labels );
     
```
